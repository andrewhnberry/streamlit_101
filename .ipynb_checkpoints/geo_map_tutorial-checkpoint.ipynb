### Getting Canadian Postal Codes to latitude and logitude

```python
#Python Library for postal code geocoding and distance calcualtion
#Very useful if working with Canadian Postal Codes
import pgeocode
nomi = pgeocode.Nominatim('ca')
postal_code = 'M5V 1Z4'

location = nomi.query_postal_code(postal_code)
print(location.latitude, location.longitude)
```

```python
def get_lat(Canadian_Postal_Code):
    
    location = nomi.query_postal_code(Canadian_Postal_Code)
    
    return location.latitude

def get_long(Canadian_Postal_Code):
    
    location = nomi.query_postal_code(Canadian_Postal_Code)
    
    return location.longitude
        
```

```python
get_long(df['SHELTER_POSTAL_CODE'][2222])
```

```python
get_lat(df['SHELTER_POSTAL_CODE'][2222])
```

```python
df['latitude'] = np.nan
df['longitude'] = np.nan
```

```python
%%time
for i in np.arange(0, len(df)):
    df['latitude'][i] = get_lat(df['SHELTER_POSTAL_CODE'][i])
    
    df['longitude'][i] = get_long(df['SHELTER_POSTAL_CODE'][i])
```

```python
df['OCCUPANCY_YEAR'] = df['OCCUPANCY_DATE'].dt.year
df['OCCUPANCY_MONTH'] = df['OCCUPANCY_DATE'].dt.month
df['OCCUPANCY_DAY'] = df['OCCUPANCY_DATE'].dt.day
```
